```python
fig = plt.figure(figsize = (15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df["data"], df["temperatura"], color = "g")
eixo.set_title("temperatura no momento", size = 25, pad = 20)
eixo.set_xlabel("Data", size = 20)
eixo.set_ylabel("Temperatura", size = 20)
eixo.legend(["Temperatura"], loc = "lower right", fontsize = 15)
eixo.grid(True)

x1 = df["data"][df["temperatura"].idxmax()] #Definindo o limite do eixo X para receber a anotação
y1 = max(df["temperatura"]) #Definindo o limite do eixo y para receber a anotação

eixo.annotate("Máximo", xy = (x1, y1), fontsize = 18) #Definindo a anotação utilizando as coordenadas de x1 e y1 

eixo.axhline(max(df["temperatura"]), color = "k", linestyle = "--")
eixo.axhline(min(df["temperatura"]), color = "k", linestyle = "--")
```

```python
fig = plt.figure(figsize = (15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df["data"], df["temperatura"], color = "g")
eixo.set_title("temperatura no momento", size = 25, pad = 20)
eixo.set_xlabel("Data", size = 20)
eixo.set_ylabel("Temperatura", size = 20)
eixo.legend(["Temperatura"], loc = "lower right", fontsize = 15)
eixo.grid(True)

x1 = df["data"][df["temperatura"].idxmax()] #Definindo o limite do eixo X para receber a anotação
y1 = max(df["temperatura"]) #Definindo o limite do eixo y para receber a anotação

x2 = df["data"][df["temperatura"].idxmax() - 7000] #Definindo o limite do eixo X para receber a seta
y2 = max(df["temperatura"]) - 5 #Definindo o limite do eixo y para receber a seta

eixo.annotate("Máximo", xy = (x1, y1), fontsize = 18,
             xytext = (x2, y2), arrowprops = dict(facecolor = "k")) #Definindo a anotação e a seta utilizando as coordenadas de x1/x2 e y1/y2 

x3 = df["data"][df["temperatura"].idxmin()] #Definindo o limite do eixo X para receber a anotação
y3 = min(df["temperatura"]) #Definindo o limite do eixo y para receber a anotação

x4 = df["data"][df["temperatura"].idxmin() - 7000 ] #Definindo o limite do eixo X para receber a seta
y4 = min(df["temperatura"]) + 5 #Definindo o limite do eixo y para receber a seta

eixo.annotate("Mínimo", xy = (x3, y3), fontsize = 18,
                             xytext = (x4, y4), arrowprops = dict(facecolor = "k")) #Definindo a anotação e a seta utilizando as coordenadas de x1/x2 e y1/y2

eixo.axhline(max(df["temperatura"]), color = "k", linestyle = "--") # Pontilhado para o máximo
eixo.axhline(min(df["temperatura"]), color = "k", linestyle = "--") # Pontilhado para o mínimo
```

```python
temperatura_dias_da_semana = df.groupby("dia_da_semana")["temperatura"].mean()
nome_dias = ["Domingo", "Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado"]
temperatura_dias_da_semana = temperatura_dias_da_semana[nome_dias] # Para a Series vir na ordem de dias da semana correta
temperatura_dias_da_semana
```

```python
fig = plt.figure(figsize = (5,4))
ax = fig.add_axes([0,0,1,1])

cores = ["k", "r", "g", "b", "yellow", "orange", "pink"]

ax.bar(nome_dias,temperatura_dias_da_semana, color = cores, edgecolor = "black")
ax.set_title("Gráfico", fontsize = 18, pad = 10)
ax.set_xlabel("Dia da semana", fontsize = 15)
ax.set_ylabel("Média de temperatura", fontsize = 15)

```

```python
temperatura_dias_da_semana = temperatura_dias_da_semana[reversed(nome_dias)]
temperatura_dias_da_semana
```

```python
fig = plt.figure(figsize = (5,4))
ax = fig.add_axes([0,0,1,1])
ax.pie(temperatura_dias_da_semana, labels = temperatura_dias_da_semana.index, autopct="%.1f%%", explode = [0.1,0,0,0.1,0,0,0.1],
      shadow = True)
ax.set_title("Dias da semana", fontweight = 1000, fontsize = 18)
ax = ax
```
