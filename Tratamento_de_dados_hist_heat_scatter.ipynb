```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("countries.csv")
df
```

```python
df.info()
```

```python
df.country.unique()
```

```python
df_brazil = df[df["country"] == "Brazil"]
df_usa = df[df["country"] == "United States"]
```

```python
df_brazil.plot(x = "year", y = "population")
df_usa.plot(x = "year", y = "population")
```

```python
plt.plot(df_brazil["year"], df_brazil["population"])
plt.plot(df_usa["year"], df_usa["population"])
plt.legend(["Brazil", "USA"])
plt.xlabel("Year")
plt.ylabel("Population")
```

```python
df_brazil["population_rel"] = df_brazil["population"] / df_brazil.iloc[0,2] 
df_usa["population_rel"] = df_usa["population"] / df_usa.iloc[0,2]
```

```python
df_brazil
```

```python
df_usa
```

```python
plt.plot(df_brazil["year"], df_brazil["population_rel"])
plt.plot(df_usa["year"], df_usa["population_rel"])
plt.legend(["Brazil", "USA"])
plt.xlabel("Year")
plt.ylabel("Population")
```

```python
usa = df[df["country"] == "United States"].copy()
br = df[df["country"] == "Brazil"].copy()
can = df[df["country"] == "Canada"].copy()
```

```python
barras = ["USA", "Brazil", "Canada"]
alturas = [usa.population.mean(), br.population.mean(), can.population.mean()]
alturas
```

```python
plt.bar(x = barras, height = alturas)
```

```python
usa["population_rel"] = usa.population / usa.iloc[0,2]
br["population_rel"] = br.population / br.iloc[0,2]
can["population_rel"] = can.population / can.iloc[0,2]
```

```python
barras = ["USA", "Brazil", "Canada"]
alturas = [usa.population_rel.mean(), br.population_rel.mean(), can.population_rel.mean()]
```

```python
plt.bar(barras, alturas)
plt.title("MÃ©dia relativa de cresimento", size = 15)
plt.ylabel("% relativo", size = 13)
```

# **Trabalhando com Histogramas e grÃ¡ficos de frequÃªncia**

```python
df2 = pd.read_csv("housing.csv")
df2.head()
```

```python
df2.info()
```

```python
df2.quartos = df2.quartos.astype("int8")
df2.vagas = df2.vagas.astype("int8")
df2.banheiros = df2.banheiros.astype("int8")
df2.info()
```

```python
quartos_freq = df2.quartos.value_counts()
quartos_freq.index
plt.bar(quartos_freq.index, quartos_freq)
plt.title("DistribuiÃ§Ã£o de nÃºmero de quartos")
```

```python
plt.hist(df2.preco)
```

Retirando os outliers do histograma do preÃ§o

```python
plt.hist(df2[df2.preco < 5e6].preco) # A notaÃ§Ã£o 5e6 equivale a escrever 5,000.000
```

```python
plt.hist(df2[df2.tipo == "casa"].preco)
```

```python
plt.hist(df2[df2.tipo == "apartamento"].preco)
```

```python
plt.hist(df2[df2.tipo == "kitnet"].preco)
```

```python
plt.figure (figsize = (20,6))
for i, tipo in enumerate(df2.tipo.unique()):
  plt.subplot(1, 3, i + 1)
  plt.title(tipo)
  plt.hist(df2[df2.tipo == tipo].preco, color = "skyblue")
```

```python
df2.columns
```

```python
plt.figure(figsize = (20,10))
plt.scatter(df2.longitude, df2.latitude, c = df2.preco, cmap = "plasma", s= df2.area_construida / 3, alpha = 0.5)
plt.axis("equal")
plt.colorbar()

# ParÃ¢metro "c" = Muda a cor dos pontos dentro de uma escala de cor utilizando uma Series como variÃ¡vel, no caso a coluna "preco"
# ParÃ¢metro "s" = Muda o tamanho dos pontos do grÃ¡fico utilizando uma Series como variÃ¡vel, no caso a coluna "area construida"
# ParÃ¢metro "alpha" = Define o nÃ­vel de transparÃªncia dos pontos para evitar sobreposiÃ§Ã£o. 
```

```python
filtro = (df2.preco < 5e6) & (df2.area_construida < 500)

plt.figure(figsize = (15,7))
plt.scatter(df2[filtro].area_construida, df2[filtro].preco, color = "orange")
plt.title("Ãrea ConstruÃ­da x PreÃ§o", size = 15)
plt.xlabel("Ãrea construÃ­da", size = 15)
plt.ylabel("PreÃ§o", size = 15)
```
