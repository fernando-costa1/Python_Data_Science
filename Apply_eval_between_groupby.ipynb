```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("housing.csv")
df.info()
```

**Criando função personalizada para aplicação em uma Series:**

```python
# Definindo se a distância até o aeroporto é "curta" ou "longa" a partir de uma condição específica

def pert_long(dist):
  if dist > 3.5:
    return "Longa"
  else:
    return "Curta"
```

```python
df["Distancia"] = df.distancia_aeroporto.apply(pert_long)
df.drop(labels = "Distância", axis = 1, inplace = True)
```

```python
df.head()
```

```python
# Criando coluna a partir de operação com outras colunas do Data Frame
df.eval("razao_area = area_construida / area_terreno", inplace = True)
df.head()
```

```python
# Filtrando a partir de distância para o aeroporto entre 1 e 3 
df[df.distancia_aeroporto.between(1,3)].head() 
```

# **Utilizando groupby:**

```python
# Média dos preços a partir dos tipos

df[['tipo', 'preco']].groupby(by = "tipo").mean()
```

```python
# Preço máximo a partir de tipo e número de quartos
df.groupby(by = ['tipo', 'quartos']).max()
```

```python
df.groupby(by = ['tipo', 'quartos', 'vagas']).max().preco.sort_values(ascending = False)
```
