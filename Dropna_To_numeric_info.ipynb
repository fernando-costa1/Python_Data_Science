```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("housing_dados_brutos.csv")
df.head()
```

```python
df.info()
```

```python
# Excluindo linhas com valores nulos
df.dropna(axis = 0, inplace = True)
df.info()
```

```python
df.columns
```

```python
# Renomeando colunas

df.rename(columns={'Bairro':'bairro', 'Endereco':'endereco', 'Quartos':'quarto', 'Tipo':'tipo', 'Preco':'preco',
       'Distancia do aeroporto':'distancia_aeroporto', 'CEP':'cep', 'Banheiro':'banheiro', 'Vagas':'vagas', 'Area terreno':'area_terreno',
       'Area construida':'area_construida', 'Ano da construcao':'ano_construcao', 'Latitude':'latitude', 'Longitude':'longitude'}, inplace = True)
```

```python
df.columns
```

```python
#Transformando as colunas de latitude e longitude em números
df.latitude = pd.to_numeric(df.latitude)
df.longitude = pd.to_numeric(df.longitude, errors = "coerce")
df.dropna(axis = 0, inplace = True)
df.info()
```

```python
 df[["latitude", "longitude"]].describe()
```

```python
# Eliminando valores inconsistentes das colunas "latitude" e "longitude"

df.loc[df.latitude < -100, "latitude"] = df.latitude / 1000
df.latitude.describe()
```

```python
df.loc[df.longitude > 200, "longitude"] = df.longitude / 1000
df.longitude.describe()
```
