# TESTE DE HIPÓTESE

```python
# Hnull = A distribuição da taxa da gorgeta é a mesma nos dois grupos
# Halt = A distribuição da taxa da gorgeta NÃO é a mesma nos dois grupos
```

```python
from scipy.stats import ranksums
```

```python
sobremesa = gorgetas.query("sobremesa == 'sim'").porcentagem
sobremesa
```

```python
sem_sobremesa = gorgetas.query("sobremesa == 'não'").porcentagem
sem_sobremesa
```

```python
teste = ranksums(sobremesa, sem_sobremesa)
```

```python
print("O valor do teste de hipótese é: {}".format(teste.pvalue))
```

```python
# resultado maior que 0,05 siginifica que a hipótese é verdadeira. 
```
